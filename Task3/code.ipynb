```python
import random
import re
import time
from IPython.display import clear_output

class AlienBot:
    # Response lists
    negative_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")

    questions = (
        "Why are you here?",
        "Are there many humans like you?",
        "What do you consume for sustenance?",
        "Is there intelligent life on this planet?",
        "Does Earth have a leader?"
    )

    def __init__(self):
        self.intents = {
            'describe_planet_intent': r'.*\s*your planet.*',
            'answer_why_intent': r'why\sare.*',
            'about_intellipaat': r'.*\s*intellipaat.*',
            'how_are_you_intent': r'.*\s*how are you.*',
            'your_name_intent': r'.*\s*your name.*'
        }

    def greet(self):
        self.user_name = input("👽 What is your name?\n")
        time.sleep(1)
        will_help = input(f"👽 Hi {self.user_name}, I am a bot. Will you help me learn about your planet?\n").lower()
        if will_help in self.negative_responses:
            print("👽 Have a nice Earth day! 🌎")
            return
        self.chat()

    def make_exit(self, user_input):
        if user_input in self.exit_commands:
            print("👽 Have a nice day! 🌟")
            return True
        return False

    def chat(self):
        user_input = input("👽 " + random.choice(self.questions) + "\n").lower()
        while not self.make_exit(user_input):
            user_input = input("👽 " + self.match_reply(user_input) + "\n").lower()

    def match_reply(self, user_input):
        for intent, pattern in self.intents.items():
            if re.match(pattern, user_input):
                return getattr(self, intent)()

        return self.no_match_intent()

    def describe_planet_intent(self):
        responses = (
            "🌌 My planet is a utopia of diverse organisms.",
            "☕ I heard the coffee is good."
        )
        return random.choice(responses)

    def answer_why_intent(self):
        responses = (
            "✌️ I come in peace.",
            "📊 I am here to collect data on your planet and its inhabitants.",
            "☕ I heard the coffee is good."
        )
        return random.choice(responses)

    def about_intellipaat(self):
        responses = (
            "💼 Intellipaat is the world's largest professional educational company.",
            "📚 Intellipaat will make you learn concepts in a unique way.",
            "🚀 Intellipaat is where your career and skills grow."
        )
        return random.choice(responses)

    def how_are_you_intent(self):
        responses = (
            "🤖 I am just a bot, but I am functioning as expected.",
            "👽 I am here to learn about your planet."
        )
        return random.choice(responses)

    def your_name_intent(self):
        responses = (
            "🤖 I am called AlienBot.",
            "👽 I am just a bot with no formal name, but you can call me AlienBot."
        )
        return random.choice(responses)

    def no_match_intent(self):
        responses = (
            "🔍 Please tell me more.",
            "📖 Tell me more!",
            "🧐 I see. Can you elaborate?",
            "🤔 Interesting. Can you tell me more?",
            "💭 I see. What do you think?",
            "❓ Why?",
            "🤔 How do you think I feel when you say that? Why?"
        )
        return random.choice(responses)

# Main function to run the bot
def run_bot():
    clear_output(wait=True)
    bot = AlienBot()
    bot.greet()

# Run the bot
run_bot()
```
