```python
import random
import re
import time
from IPython.display import clear_output

class AlienBot:
    # Response lists
    negative_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")

    questions = (
        "Why are you here?",
        "Are there many humans like you?",
        "What do you consume for sustenance?",
        "Is there intelligent life on this planet?",
        "Does Earth have a leader?"
    )

    def __init__(self):
        self.intents = {
            'describe_planet_intent': r'.*\s*your planet.*',
            'answer_why_intent': r'why\sare.*',
            'about_intellipaat': r'.*\s*intellipaat.*',
            'how_are_you_intent': r'.*\s*how are you.*',
            'your_name_intent': r'.*\s*your name.*'
        }

    def greet(self):
        self.user_name = input("ğŸ‘½ What is your name?\n")
        time.sleep(1)
        will_help = input(f"ğŸ‘½ Hi {self.user_name}, I am a bot. Will you help me learn about your planet?\n").lower()
        if will_help in self.negative_responses:
            print("ğŸ‘½ Have a nice Earth day! ğŸŒ")
            return
        self.chat()

    def make_exit(self, user_input):
        if user_input in self.exit_commands:
            print("ğŸ‘½ Have a nice day! ğŸŒŸ")
            return True
        return False

    def chat(self):
        user_input = input("ğŸ‘½ " + random.choice(self.questions) + "\n").lower()
        while not self.make_exit(user_input):
            user_input = input("ğŸ‘½ " + self.match_reply(user_input) + "\n").lower()

    def match_reply(self, user_input):
        for intent, pattern in self.intents.items():
            if re.match(pattern, user_input):
                return getattr(self, intent)()

        return self.no_match_intent()

    def describe_planet_intent(self):
        responses = (
            "ğŸŒŒ My planet is a utopia of diverse organisms.",
            "â˜• I heard the coffee is good."
        )
        return random.choice(responses)

    def answer_why_intent(self):
        responses = (
            "âœŒï¸ I come in peace.",
            "ğŸ“Š I am here to collect data on your planet and its inhabitants.",
            "â˜• I heard the coffee is good."
        )
        return random.choice(responses)

    def about_intellipaat(self):
        responses = (
            "ğŸ’¼ Intellipaat is the world's largest professional educational company.",
            "ğŸ“š Intellipaat will make you learn concepts in a unique way.",
            "ğŸš€ Intellipaat is where your career and skills grow."
        )
        return random.choice(responses)

    def how_are_you_intent(self):
        responses = (
            "ğŸ¤– I am just a bot, but I am functioning as expected.",
            "ğŸ‘½ I am here to learn about your planet."
        )
        return random.choice(responses)

    def your_name_intent(self):
        responses = (
            "ğŸ¤– I am called AlienBot.",
            "ğŸ‘½ I am just a bot with no formal name, but you can call me AlienBot."
        )
        return random.choice(responses)

    def no_match_intent(self):
        responses = (
            "ğŸ” Please tell me more.",
            "ğŸ“– Tell me more!",
            "ğŸ§ I see. Can you elaborate?",
            "ğŸ¤” Interesting. Can you tell me more?",
            "ğŸ’­ I see. What do you think?",
            "â“ Why?",
            "ğŸ¤” How do you think I feel when you say that? Why?"
        )
        return random.choice(responses)

# Main function to run the bot
def run_bot():
    clear_output(wait=True)
    bot = AlienBot()
    bot.greet()

# Run the bot
run_bot()
```
